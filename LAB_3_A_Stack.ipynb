```python
import ctypes
class Stack:
    def __init__(self, size):
        self.size = size 
        self.top  = - 1
        MyArray = ctypes.py_object*size
        self.data = MyArray()
        
    def push(self, value):
        self.top = self.top + 1
        self.data[self.top] = value
        
    def pop(self):
        if self.is_empty() == 1:
            print("Stack is Empty")
        else:
            value = self.data[self.top]
            self.top = self.top - 1
            return value
            
    def peek(self):
        if self.is_empty() == 1:
            print("Stack is Empty")
        
        else:
            print("The uppermost value of the stack :", self.data[self.top])
        
    def is_empty(self):
        if self.top == -1:
            return 1
        
    def is_full(self):
        if self.top == self.size - 1:
            return 1
        
    def display(self):
        if self.is_empty() == 1:
            print("Stack is Empty")
        else:
            for i in range(0, self.size - 1):
                print(self.data[i])
            
    
        
def main():
    size = int(input("Enter the size of stack: "))
    ob = Stack(size)
    
    print("1 - Push")
    print("2 - Pop")
    print("3 - Peek")
    print("4 - Display")
    print("5 - Exit")
    
    while True:
        ch = int(input("Enter your option: "))
        if ch == 5:
            print("Exit")
            return
        if ch == 1:
            if ob.is_full == 1:
                print("Stack is full")
            else:
                value = int(input("Enter the value you want to push in the Stack: "))
                ob.push(value)
                
        if ch == 2:
            value = ob.pop()
            print(value, "has been popped")
            
        if ch == 3:
            ob.peek()
        if ch == 4:
            ob.display()
    
    
main()    
    
    
```
