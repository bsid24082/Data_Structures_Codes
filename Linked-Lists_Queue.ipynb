```python
class Node:
    def __init__(self,data):
        self.data = data
        self.ref = None
        
class Queue:
    def __init__(self):
        self.front = self.rear = None
        
    def isEmpty():
        return self.front == None
   
    
    def Insert(self, item):
        new_node = Node(item)
        
        if self.rear == None:
            self.front = self.rear = new_node
        else:
            self.rear.ref = new_node
            self.rear = new_node
    def Delete(self):
        
        if self.front == None:
            print("Queue is Empty")
            
        else:
            value = self.front.data
            self.front = self.front.ref
            print("Delete - ", value)
            
        if(self.front == None):
            self.rear = None
            
    def Display(self):
        if self.front == None:
            print("Queue is Empty")
            
        else:
            temp = self.front
            while temp is not None:
                print(temp.data)
                temp = temp.ref
                
                
                
def main():
    queue = Queue()
    
    
    print("1- Insert")
    print("2- Delete")
    print("3- Display")
    print("0- Exit")
    
    
    while True:
        ch = int(input("Enter your choice: "))
        if ch == 0:
            print("Exit")
            break;
        if ch == 1:
            value = int(input("Enter the value to be Inserted: "))
            queue.Insert(value)
        if ch == 2:
            queue.Delete()
            
        if ch == 3:
            queue.Display()
            
main()            
    
```
