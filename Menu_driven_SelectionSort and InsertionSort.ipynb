# Menu-Driven Insertion and Selection Sort #

```python
def selection_sorting(array):
    for i in range(len(array)):
        # find the minimum element in remaining 
        # unsorted array

        min_idx = i
        for j in range(i+1,len(array)):
            if array[min_idx] > array[j]:
                min_idx = j
        array[i],array[min_idx] = array[min_idx],array[i]
        
def InsertionSort(arr):
    ## Transverse through 1 to len(arr)
    for i in range(1, len(arr)):
        key = arr[i]
        ## Move Elements of arr[0...i-1], that are 
        ## greater than key, to one position ahead
        ## of their currentr position 
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
        
def main():
    while True:
        print("------- Select the Sorting method -------")
        print("1. Selection Sorting")
        print("2. Insertion Sorting")
        print("Exit")
        
        ch = int(input("Enter your choice: "))
       
        if ch == 1:
            array = []
            a = int(input("Enter the number of elements: "))
            for i in range(a):
                a = int(input("Enter the values: "))
                array.append(a)
            selection_sorting(array)
            
            print("Sorted array is: ",array)
            
        elif ch == 2:
            arr = []
            a  = int(input("Enter no. of elements you want: "))
            for i in range(0, a):
                ele = int(input())
                arr.append(ele)
            InsertionSort(arr)
            print(arr)
        else:
            print("Exiting the program")
            break
            
main()
```

```python
def Selection_Sort(array):
    for i in range(len(array)):
        min_idx = i
        for j in range(i+1, len(array)):
            if array[min_idx] > array[j]:
                min_idx = j 
        array[i], array[min_idx] = array[min_idx], array[i] 
    
def Insertion_Sort(array):
    for i in range(1, len(array)):
        key = array[i]
        j = i - 1
        
        while j >= 0 and key < array[j]:
            array[j+1] = array[j]
            j = j - 1
        array[j + 1] = key    
def main():
    while True:
        print("1 - Selection-Sort")
        print("2 - Insertion Sort")
        print("0 - Exit")
        ch = int(input("Enter your choice: "))
        if ch == 0:
            print("Exit")
            break
        if ch == 1:
            print("Selection Sort")
            array = []
            a = int(input("Elements..."))
            for i in range(a):
                ele = int(input("Enter the elements: "))
                array.append(ele)
            Selection_Sort(array)
            print(array)
        if ch == 2:
            print("Insertion Sort")
            array = []
            a = int(input("Elements..."))
            for i in range(a):
                ele = int(input("Enter the elements: "))
                array.append(ele)
            Insertion_Sort(array)
        
            print(array)
        
main()
        
    
```
