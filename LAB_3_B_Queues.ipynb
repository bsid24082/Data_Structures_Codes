```python
import ctypes

class Queue:
    #Create a queue of given size 
    def __init__(self,size):
        self.size = size
        self.front = -1
        self.rear = -1
        Myarray = ctypes.py_object*size
        self.data = Myarray()

    def is_empty(self):
        if self.front == -1:
            return 1
        else:
            return 0
            
    def is_full(self):
        if self.rear==self.size-1:
            return 1
        else:
            return 0
            
    def insert(self,value):
        if self.front == -1:
            self.front = 0
        self.rear = self.rear + 1
        self.data[self.rear] = value
        print('Value is inserted')
    
    def delete(self):
        if self.is_empty():
            print('The queue is empty')
        else:
            value = self.data[self.front] #Popping deleted value and storing in the variable 'value' which will be shown to user
            self.front = self.front + 1
            if self.front>self.rear: #Queue has become empty 
                self.front = self.rear = -1
            return value 
    
    def display(self):
        if self.is_empty():
            print('Queue is empty')
        else:
            for i in range(self.front, self.rear+1):
                print(self.data[i])
                
                
def main():
    size = int(input('Enter size of Queue: '))
    ob = Queue(size)
    print('1. Insert')
    print('2. Delete')
    print('3. Display')
    print('0. Exit')
    
    while True:
        ch = int(input("Enter your choice: "))
        
        if ch==0:
            print('Exit')
            break
        
        elif ch==1:
            if ob.is_full():
                print('The Queue is full')
            else:
                value = int(input("Enter the value to be inserted: "))
                ob.insert(value)
                
        elif ch==2:
                value = ob.delete()
                print(value,' value is deleted from the stack')
        elif ch==3:
            ob.display()
        else:
            print('The choice entered is wrong')
main()     
    
```
