# Linked-List (Stacks)

```python
class Node:
    def __init__(self,data):
        self.item = data
        self.ref = None
        
class LinkedList:
    def __init__(self):
        self.start_node = None

    def insert_at_end(self,data):
        new_node = Node(data)
        if self.start_node is None:
            self.start_node = new_node
            return
        temp = self.start_node
        while temp.ref is not None:
            temp = temp.ref
            temp.ref = new_node
            
    def display(self):
        if self.start_node == None:
            print("Empty")
            return
        else:
            temp = self.start_node
            while temp is not None:
                print(temp.item)
                temp = temp.ref
    
    def insert_at_first(self,value):
        new_node=Node(value)
        new_node.ref=self.start_node
        self.start_node=new_node
       
    
    
    def insert_at_pos(self,value,pos):
        new_node=Node(value)
        if pos==1:
            self.insert_at_first(value)
            return
        temp=self.start_node
        for i in range(1, pos-1):
            temp = temp.ref
            if(temp == None):
                print("Wrong Index")
                return
        new_node.ref = temp.ref
        temp.ref = new_node
        
    def delete_at_first(self):
        if self.start_node == None:
            print("List is Empty")
        else:
            value = self.start_node.item
            self.start_node = self.start_node.ref
            print("Element is deleted", value)
            
    def delete_at_last(self):
        if self.start_node == None:
            print("List is Empty")
        else:
            temp = self.start_node
            while temp.ref.ref is not None:
                temp = temp.ref
            value = temp.ref.item
            temp.ref = None
            print("Value is deleted",value)
            
    def delete_at_pos(self,pos):
        if self.start_node==None:
            print("List is empty")
        if pos==1:
            self.delete_at_first()
        else:
            temp=self.start_node
            for i in range(1,pos-1):
                temp=temp.ref
                if(temp==None):
                    print("Wrong Index")
                    return
            value=temp.ref.item
            temp.ref=temp.ref.ref
            print("Value is deleted",value)
            
def main():
    stack=LinkedList()
    
    print("1.Insert in the end")
    print("2.Insert in the beginning")
    print("3.Display")
    print("4.Insert at given position")
    print("5.Delete at last")
    print("6.Delete at first")
    print("7.Delete at given position")
    print("0.Exit")
    
    while True:
        ch=int(input("Enter your choice"))
        if ch==0:
            print("Exit")
            break
        elif ch==1:
            value=int(input("Enter the value to be inserted-"))
            stack.insert_at_end(value)
        elif ch==2:
            value=int(input("Enter the value to be inserted-"))
            stack.insert_at_first(value)
        elif ch==3:
            stack.display()
        elif ch==4:
            value=int(input("Enter the value to be inserted-"))
            pos=int(input("Enter the position you want to enter at"))
            stack.insert_at_pos(value,pos)
        elif ch==5:
            stack.delete_at_last()
        elif ch==6:
            stack.delete_at_first()
        elif ch==7:
            pos=int(input("Enter the position you want to delete at"))
            stack.delete_at_pos(pos)
        else:
            print("Invalid choice")
main()
```
