```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class circular_linked_list:
    def __init__(self):
        self.start = None
    
    def insert_at_last(self,value):
        new_node = Node(value)
        if self.start == None:
            self.start = new_node
            new_node.next = self.start
        else:
            temp = self.start
            while temp.next != self.start:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.start
            print("Value is Inserted")
            
    def insert_at_start(self, value):
        new_node = Node(value)
        if(self.start == None):
            self.start = new_node
            new_node.next = self.start
        else:
            temp = self.start
            while temp.next != self.start:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.start
            self.start = new_node
            
    def insert_at_pos(self, value,pos):
        new_node = Node(value)
        if(self.start == None):
            self.start = new_node
            new_node.next = self.start
        if (pos ==1):
            self.insert_at_start(value)
        else:
            temp = self.start
            for i in range(1,pos-1):
                temp = temp.next
                if(temp==None):
                    print("Invalid Index")
            new_node.next = temp.next
            temp.next = new_node
            
    def remove_first(self):
        if self.start == None:
            print("List is Empty")
        else:
            temp = self.start 
            while temp.next != self.start:
                temp = temp.next
            temp.next = self.start.next
            value = self.start.data
            self.start = self.start.next
            print(value, "is Deleted")
            
    def remove_last(self):
        if self.start == None:
            print("List is Empty")
        else:
            temp = self.start
            while temp.next.next != self.start:
                temp = temp.next
            value = temp.next.data
            temp.next = self.start
            print(value, "is Deleted")
      
    def remove_at_pos(self, pos):
        if self.start == None:
            print("List is Empty")
        elif pos == 1:
            self.remove_first()
        else:
            temp = self.start
            for i in range (1,pos-1):
                temp = temp.next
                if temp.next == self.start:
                    print("Invalid Index")
                    return
            value = temp.next.data
            temp.next = temp.next.next
            print(value, "is Deleted")
       
    def display (self):
        if self.start == None:
            print("List is Empty")
        else:
            temp = self.start
            while True:
                print(temp.data)
                temp = temp.next
                if temp==self.start:
                    break
    
def main():
    CLL = circular_linked_list()
    
    print("1-Insert at Last")
    print("2-Insert at Start")
    print("3-Insert at Position") 
    print("4-Delete at Last")
    print("5-Delete at Start")
    print("6-Delete at Position")
    print("7-Display")
    print("0-Exit")

    while True:
        ch = int(input("ENTER YOUR CHOICE-"))
        if ch == 0:
            print("EXIT")
            break;
        if ch==1:
            value = int(input("ENTER THE VALUE-"))
            CLL.insert_at_last(value)
        if ch==2:
            value = int(input("ENTER THE VALUE-"))
            CLL.insert_at_start(value)    
        if ch==3:
            value = int(input("ENTER THE VALUE-"))
            pos = int(input("ENTER THE POSITION-"))
            CLL.insert_at_pos(value,pos)
        if ch==4:                    
            CLL.remove_last()
        if ch==5:                    
            CLL.remove_first()
        if ch==6:
            pos = int(input("ENTER THE POSITION-"))
            CLL.remove_at_pos(pos)
        if ch==7:
            print("LIST CONTENTS")
            CLL.display()
        
        
main()  

```
