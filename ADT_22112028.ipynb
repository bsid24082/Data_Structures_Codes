```python
import ctypes
#Create an array and initialise it
class Array:
    def __init__(self,size):
        self.size = size
        Myarray = ctypes.py_object*size
        self.elements = Myarray()
        for i in range(self.size-1):
            self.elements[i] = None
            
#Set the initial values in Array
    def __set__(self):
        x = 10
        for i in range(self.size-1):
            self.elements[i] = x
            x += 10
            
#Display Array contents
    def __display__(self):
        for i in range(0,self.size-1):
            print(self.elements[i])
    
#Insert given value at given index.Value will not be inserted
    def __insert_at_index__(self,index,value):
        if self.elements[index] == None:
            self.elements[index] = value
        else:
            print("No empty location")
    
#Reset an Array to None
    def __clear__(self):
        for i in range(0,self.size-1):
            self.elements[i] = None
#Delete the given value
    def __delete_by_value__(self,value):
        x=0
        for i in range(0,self.size-1):
            if self.elements[i] == value:
                self.elements[i] = None
                x = 1
                print("Value is deleted")
        if x == 0:
            print('Element does not exist')
#Search the given value in an array
    def __search__(self,value):
        x = 0
        for i in range(0,self.size-1):
            if self.elements[i] == value:
                print("Value is present at",i+1)
                x = 1
        if x == 0:
            print("Element does not exist")
#Update the given value at given Index
    def update(self,index,value):
        if index >= 0 and index < self.size-1:
            self.elements[index] = value
        else:
            print("Invalid index")
#Count and returns total None values from an array
    def count(self):
        cnt = 0
        for i in range(0,self.size-1):
            if self.elements[i] != None:
                cnt = cnt + 1
        return cnt
#Return TRUE(1) if array is empty else returns False(0)
    def is_empty(self):
        if self.count() == 0:
            return 1
        else:
            return 0
#Return TRUE(1) if array is full else returns False(0)
    def is_full(self):
        if self.count() == self.size-1:
            return 1
        else:
            return 0

#if __name__ =="__main__":
def main():
    size = int(input("Enter size of an array"))
    ob = Array(size+1)
    
    while True:
        print("1.Set an array")
        print("2.Insert at Index")
        print("3.Display")
        print("4.Reset")
        print("5.Delete by Value")
        print("6.Update the value at given Index")
        print("7.Count Elements")
        print("8.Is_full")
        print("9.Is_Empty")
        print("0.Exit")
        ch = int(input("Enter your choice"))
        if ch == 0:
            print("Exit")
            break;
        elif ch == 1:
            ob.__set__()
            print("Array is created")
        elif ch == 2:
            value = int(input("Enter value to be inserted"))
            index = int(input("Enter index where value to be inserted"))
            ob.__insert_at_index__(index,value)
        elif ch == 3:
            ob.__display__()
        elif ch == 4:
            ob.__clear__()
        elif ch == 5:
            value = int(input("Enter value to be deleted"))
            ob.__delete_by_value__(value)
        elif ch == 6:
            index = int(input("Enter index"))
            value = int(input("Enter value"))
            ob.update(index,value)
        elif ch == 7:
            print("Total no of elements=",ob.count())
        elif ch == 8:
            if ob.is_full() == 1:
                print("Array is full")
            else:
                print("Array is not full")
        elif ch == 9:
            if ob.is_empty() == 1:
                print("Array is empty")
            else:
                print("Array is not empty")
        else:
            print("Wrong choice entered")
    
main()
```
